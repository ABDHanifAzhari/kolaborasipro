```python
import pandas as pd
import numpy as np
from numpy import array
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("airline-passengers.csv")
data
```

```python
passengers=data["Passengers"]
passengers
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test = train_test_split(passengers, test_size=0.2)
```

```python
scaler = MinMaxScaler()
scaled = scaler.fit_transform(np.reshape(X_train,(-1,1)))
scaled_features = pd.DataFrame(scaled)
scaled_features
```

```python
# split a univariate sequence into samples
def split_sequence(sequence, n_steps):
  X, y = list(), list()
  for i in range(len(sequence)):
    # find the end of this pattern
    end_ix = i + n_steps
    # check if we are beyond the sequence
    if end_ix > len(sequence)-1:
      break
    # gather input and output parts of the pattern
    seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
    X.append(seq_x)
    y.append(seq_y)
  return array(X), array(y)
```

```python
# transform to a supervised learning problem
X, y = split_sequence(scaled_features[0].values, 3)
print(X.shape, y.shape)
dfX = pd.DataFrame(X, columns=["Xt-3","Xt-2", "Xt-1"])
dfy = pd.DataFrame(y, columns=["Xt"])

df = pd.concat((dfX, dfy), axis = 1)
df
```
